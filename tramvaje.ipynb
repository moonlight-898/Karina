```

trams = [
    ("Tatra T1",              60, 1951),
    ("Tatra T2",              65, 1955),
    ("Tatra T3",              65, 1960),
    ("Tatra K2",              65, 1966),
    ("Tatra T3SUCS",          65, 1976),
    ("Tatra T3M",             65, 1976),
    ("Tatra KT4D (ex-DDR)",   65, 1976),
    ("Tatra KT8D5",           65, 1984),
    ("Tatra T6A5",            65, 1991),
    ("RT6N1",                 60, 1993),
    ("Tatra T3R.P",           65, 2000),
    ("VarioLF",               65, 2003),
    ("Tatra K2S",             65, 2003),
    ("Tatra KT8D5.RN2P",      65, 2005),
    ("Škoda 14T",             60, 2005),
    ("Škoda 15T",             60, 2009),
    ("VarioLF2",              65, 2009),
    ("Pragoimex EVO1",        50, 2015),
    ("Pragoimex EVO2",        60, 2015),
]

print(f"Počet typů tramvají v databázi: {len(trams)}\n")
def treti_polozka(x):
    return x[2]

trams_sorted = sorted(trams, key=lambda x: x[2])

print("Tramvaje seřazené podle roku výroby:")
for typ, rychlost, rok in trams_sorted:
    print(f"{rok:4d} | {rychlost:2d} km/h | {typ}")
print()
moderni = [t for t in trams if t[2] >= 2000]

print("Moderní tramvaje (rok ≥ 2000):")
for typ, rychlost, rok in moderni:
    print(f" • {typ.ljust(25)} | {rok} | {rychlost} km/h")
print()
rychlosti = [t[1] for t in trams]
prumerna_rychlost = sum(rychlosti) / len(rychlosti)

print(f"Průměrná maximální rychlost: {prumerna_rychlost:.1f} km/h")
print(f"Nejvyšší rychlost:           {max(rychlosti)} km/h")
print(f"Nejnižší rychlost:           {min(rychlosti)} km/h")
print(f"Nejčastější rychlost:        {max(set(rychlosti), key=rychlosti.count)} km/h")
import matplotlib.pyplot as plt

# Nastavení české fontu (pokud máš nainstalovanou, např. Arial nebo DejaVu Sans)
plt.rcParams['font.family'] = 'DejaVu Sans'

roky = [t[2] for t in trams]

plt.figure(figsize=(12, 6))
plt.hist(roky, bins=[1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020],
         edgecolor='black', color='#d81b60', alpha=0.8)

plt.title("Rozložení pražských tramvají podle roku prvního nasazení", fontsize=16, pad=15)
plt.xlabel("Rok", fontsize=12)
plt.ylabel("Počet typů tramvají", fontsize=12)
plt.xticks([1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020])
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Přidáme počty nad sloupce
counts, bins, patches = plt.hist(roky, bins=[1950,1960,1970,1980,1990,2000,2010,2020],
                                 edgecolor='black', color='#d81b60')
for i, count in enumerate(counts):
    if count > 0:
        plt.text(bins[i] + 5, count + 0.1, str(int(count)), ha='center', fontsize=11)

plt.tight_layout()
plt.show()
from tabulate import tabulate

tabulka = sorted(trams, key=lambda x: x[2])
print(tabulate(tabulka, headers=["Typ tramvaje", "Max rychlost (km/h)", "Rok"], tablefmt="pretty"))
```
