```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def sound(self):
    pass


class Dog(Animal):
  def sound(self):
    return "haf haf"

class Cat(Animal):
  def sound(self):
    return "meow meow"

mydog = Dog("Jerry", 12)
mycat = Cat("Tom", 7)

print(f"{mydog.name} says {mydog.sound()}")
#mycat.food()
print(f"{mycat.name} says {mycat.sound()}")
```
